```python
def rgb2gray(rgb):
    r,g,b=rgb[:,:,0],rgb[:,:,1],rgb[:,:,2]
    gray=(r+g+b)/3
    return gray
```

```python
gray=rgb2gray(anime)
print(gray.shape)
```

```python
gray
```

```python
plt.imshow(gray)
```

```python
gray.min(),gray.max()
```
